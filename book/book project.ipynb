```python
books = books[['title', 'authors', 'categories', 'description', 'thumbnail']]
```

```python
books['authors'] = books['authors'].fillna('')
books['categories'] = books['categories'].fillna('')
books['description'] = books['description'].fillna('')
books['thumbnail'] = books['thumbnail'].fillna('')
```

```python
books['tags'] = (
    books['authors'] + " " +
    books['categories'] + " " +
    books['description']
)
```

```python
books['tags'] = books['tags'].apply(lambda x: x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(books['tags']).toarray()
```

```python
import pickle
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)

with open('books.pkl', 'wb') as f:
    pickle.dump(books, f)

with open('similarity_books.pkl', 'wb') as f:
    pickle.dump(similarity, f)

print(" Training complete! Files saved.")
```
